## 1. Algoritmos: la idea central (10 min)

Un **algoritmo** es una secuencia **finita y ordenada de pasos precisos** que transforma **entradas** en **salidas**. Una receta de cocina es un algoritmo: ingredientes (entrada) → pasos (proceso) → plato (salida).

Todo algoritmo bien planteado cumple cuatro propiedades:

| Propiedad | Significado |
|---|---|
| Preciso | Cada paso está definido sin ambigüedad |
| Finito | Termina en un número limitado de pasos |
| Ordenado | El orden de los pasos importa |
| Efectivo | Produce el resultado esperado |

Antes de escribir código conviene pensar el algoritmo en **pseudocódigo** (lenguaje natural estructurado):

```
Entrada:  tres calificaciones
Proceso:  sumarlas y dividir el total para 3
Salida:   el promedio
```

Programar es, esencialmente, **traducir algoritmos a un lenguaje que la computadora entienda**. Veamos esa traducción a Python.

````python
# 🧑‍🏫 Yo hago — Algoritmo: promedio de tres calificaciones
# Entrada:  tres calificaciones
# Proceso:  sumarlas y dividir el total para 3
# Salida:   el promedio

nota_1 = 8.5
nota_2 = 9.0
nota_3 = 7.5

suma = nota_1 + nota_2 + nota_3
promedio = suma / 3

print("El promedio es:", promedio)
````

## 2. Variables y tipos de datos (20 min)

Una **variable** es un nombre que guarda un valor. Se crea con el operador de asignación `=` (leer como "toma el valor de", no como igualdad matemática).

Reglas para nombrar variables:

| Regla | ✅ Correcto | ❌ Incorrecto |
|---|---|---|
| Sin espacios: usar `snake_case` | `tasa_interes` | `tasa interes` |
| No empezar con número | `anio_2026` | `2026_anio` |
| Sin tildes ni ñ (recomendado) | `anio` | `año` |
| Distingue mayúsculas | `pib` y `PIB` son variables distintas | — |

Los cuatro **tipos básicos**:

| Tipo | Nombre | Ejemplo |
|---|---|---|
| `int` | Entero | `18_000_000` |
| `float` | Decimal | `4.7` |
| `str` | Texto (*string*) | `"Ecuador"` |
| `bool` | Lógico | `True`, `False` |

La función `type()` revela el tipo de cualquier valor.

````python
# 🧑‍🏫 Yo hago — Variables para describir un país
pais = "Ecuador"                # str: texto, siempre entre comillas
poblacion = 18_000_000          # int: los guiones bajos solo mejoran la lectura
superficie_km2 = 256_370        # int
es_sudamericano = True          # bool: True o False (con mayúscula inicial)

densidad = poblacion / superficie_km2   # el resultado de una división es float

print(type(pais), type(poblacion), type(densidad), type(es_sudamericano))

# f-string: la forma moderna de combinar texto y variables → f"...{variable}..."
print(f"{pais} tiene {poblacion:,} habitantes y una densidad de {densidad:.1f} hab/km².")
````

En el `f-string` anterior, `{poblacion:,}` agrega separadores de miles y `{densidad:.1f}` redondea a 1 decimal.

**Conversión de tipos.** En la práctica los datos suelen llegar como texto (formularios, archivos CSV). Para operar con ellos hay que convertirlos explícitamente con `int()`, `float()` y `str()`.

````python
# Conversión explícita de tipos
anio_texto = "2026"          # esto es texto, no un número: no se puede sumar
anio = int(anio_texto)       # ahora sí es un entero
print(anio + 4)

pib_texto = "121000.5"
pib_millones = float(pib_texto)
print(pib_millones * 2)

# De número a texto (para concatenar con +)
codigo = "EC-" + str(593)
print(codigo)

# round() controla los decimales de un float
print(round(3.14159, 2))
````

## 3. Operadores (15 min)

**Aritméticos:**

| Operador | Operación | Ejemplo | Resultado |
|---|---|---|---|
| `+` `-` `*` | Suma, resta, multiplicación | `3 * 4` | `12` |
| `/` | División (siempre devuelve `float`) | `7 / 2` | `3.5` |
| `//` | División entera | `7 // 2` | `3` |
| `%` | Residuo (módulo) | `7 % 2` | `1` |
| `**` | Potencia | `2 ** 3` | `8` |

**De comparación** (el resultado siempre es un `bool`): `==` igual, `!=` distinto, `<`, `>`, `<=`, `>=`. Cuidado: `==` compara, `=` asigna.

**Lógicos** (combinan condiciones): `and`, `or`, `not`.

````python
# 🧑‍🏫 Yo hago — Operadores aritméticos en un caso económico
pib = 121_000_000_000        # PIB aproximado, en USD
poblacion = 18_000_000

pib_pc = pib / poblacion
print(f"PIB per cápita: {pib_pc:,.0f} USD")

# División entera y residuo: repartir 127 estudiantes en grupos de 25
estudiantes = 127
print("Grupos completos:", estudiantes // 25)
print("Estudiantes sobrantes:", estudiantes % 25)

# Potencia: PIB proyectado con crecimiento del 3% anual durante 5 años
pib_futuro = pib * (1 + 0.03) ** 5
print(f"PIB en 5 años: {pib_futuro:,.0f} USD")
````

````python
# 🧑‍🏫 Yo hago — Comparación y lógicos
print(pib_pc > 10_000)          # False
print(poblacion == 18_000_000)  # True  (== compara, = asigna)
print(pib_pc != 0)              # True

# and: ambas condiciones deben cumplirse
es_ingreso_medio = pib_pc > 4_500 and pib_pc < 14_000
print("¿Ingreso medio-alto?", es_ingreso_medio)

# El algoritmo del Tú haces 1, ahora en código:
numero = 7
es_par = numero % 2 == 0
print(f"¿{numero} es par? {es_par}")
print(f"¿{numero} es impar? {not es_par}")
````

## 4. Estructuras de datos básicas (20 min)

Hasta ahora cada variable guarda **un** valor. Las estructuras de datos guardan **colecciones**.

**Listas** (`[ ]`): colecciones **ordenadas y modificables**. Regla de oro: **el índice empieza en 0**, no en 1.

````python
# 🧑‍🏫 Yo hago — Listas
provincias = ["Pichincha", "Guayas", "Azuay", "Manabí"]

print(provincias[0])       # primer elemento (índice 0)
print(provincias[-1])      # último elemento (índices negativos cuentan desde el final)
print(provincias[1:3])     # rebanada: desde el índice 1 hasta ANTES del 3

provincias.append("Loja")  # agregar al final
print(provincias)

print("Elementos en la lista:", len(provincias))
print("¿Está Guayas?", "Guayas" in provincias)
````

**Tuplas** (`( )`): como las listas, pero **inmutables** (no se pueden modificar tras crearse). Útiles para datos que no deben cambiar: coordenadas, fechas, pares fijos.

**Diccionarios** (`{ }`): pares **clave → valor**. Ideales para datos etiquetados: en lugar de recordar posiciones, accedes por nombre.

````python
# 🧑‍🏫 Yo hago — Tuplas y diccionarios
coordenadas_quito = (-0.1807, -78.4678)   # (latitud, longitud)
print("Latitud de Quito:", coordenadas_quito[0])

poblacion_provincial = {
    "Pichincha": 3_200_000,
    "Guayas": 4_500_000,
    "Azuay": 900_000,
}   # cifras aproximadas, con fines didácticos

print(poblacion_provincial["Guayas"])        # acceso por clave, no por posición
poblacion_provincial["Manabí"] = 1_600_000   # agregar un par nuevo

print(list(poblacion_provincial.keys()))     # todas las claves
print(len(poblacion_provincial), "provincias registradas")
````

## 5. Estructuras de control (25 min)

Hasta aquí el código se ejecuta línea por línea, siempre igual. Las estructuras de control permiten **decidir** (`if`) y **repetir** (`for`, `while`).

### 5.1 Condicionales: `if` / `elif` / `else`

```
if condición_1:
    # se ejecuta si condición_1 es True
elif condición_2:
    # se ejecuta si condición_1 fue False y condición_2 es True
else:
    # se ejecuta si ninguna condición anterior fue True
```

Dos detalles de sintaxis que causan la mayoría de errores del día: los **dos puntos `:`** al final de la línea y la **indentación de 4 espacios**, que en Python define qué líneas pertenecen al bloque.

````python
# 🧑‍🏫 Yo hago — Clasificar un país por ingreso (umbrales referenciales del Banco Mundial)
pib_pc = 6_700   # PIB per cápita de Ecuador, USD aprox.

if pib_pc <= 1_100:
    categoria = "ingreso bajo"
elif pib_pc <= 4_500:
    categoria = "ingreso medio-bajo"
elif pib_pc <= 14_000:
    categoria = "ingreso medio-alto"
else:
    categoria = "ingreso alto"

print(f"Con {pib_pc:,} USD per cápita, el país se clasifica como {categoria}.")
````

### 5.2 Bucles `for`: repetir para cada elemento

El `for` recorre una colección y ejecuta el bloque **una vez por elemento**. Con `range(a, b)` genera la secuencia de enteros desde `a` hasta `b - 1`.

````python
# 🧑‍🏫 Yo hago — for sobre listas, rangos y diccionarios
for provincia in provincias:
    print("Provincia:", provincia)

# Patrón acumulador: sumar 1 + 2 + ... + 10
suma = 0
for i in range(1, 11):
    suma = suma + i
print("Suma de 1 a 10:", suma)

# Recorrer un diccionario obteniendo clave y valor a la vez
total = 0
for provincia, habitantes in poblacion_provincial.items():
    print(f"{provincia}: {habitantes:,} habitantes")
    total += habitantes   # abreviatura de: total = total + habitantes
print(f"Población total registrada: {total:,}")
````

### 5.3 Bucles `while`: repetir mientras se cumpla una condición

Se usa cuando **no sabemos de antemano cuántas repeticiones** hacen falta, solo la condición de parada. Advertencia: si dentro del bucle nada cambia la condición, el bucle **nunca termina** (bucle infinito). Si te ocurre, detén la ejecución con el botón ⏹ (*Interrupt*) de VS Code.

````python
# 🧑‍🏫 Yo hago — while: ¿en cuántos años se duplica el PIB creciendo al 3% anual?
pib_actual = 121.0            # miles de millones de USD
objetivo = 2 * pib_actual
anios = 0

while pib_actual < objetivo:
    pib_actual = pib_actual * 1.03   # crece 3% cada año
    anios += 1                       # si olvidas esta línea: ¡bucle infinito!

print(f"El PIB se duplica en {anios} años (la 'regla del 70' predice ≈ 70/3 ≈ {70/3:.0f}).")
````

## 6. Funciones (15 min)

Una **función** empaqueta un bloque de código con nombre propio para **reutilizarlo** sin copiar y pegar. Anatomía:

```
def nombre_funcion(parametro_1, parametro_2):
    """Docstring: qué hace la función."""
    resultado = ...       # cuerpo (indentado)
    return resultado      # valor que la función devuelve
```

Distinción clave para principiantes: `print()` solo **muestra** en pantalla; `return` **devuelve** un valor que puede guardarse en una variable y seguir usándose. Una función sin `return` devuelve `None`.

````python
# 🧑‍🏫 Yo hago — Definir y reutilizar funciones
def pib_per_capita(pib, poblacion):
    """Calcula el PIB per cápita en USD."""
    return pib / poblacion

def clasificar_ingreso(pib_pc):
    """Clasifica un país según su PIB per cápita (umbrales referenciales)."""
    if pib_pc <= 1_100:
        return "ingreso bajo"
    elif pib_pc <= 4_500:
        return "ingreso medio-bajo"
    elif pib_pc <= 14_000:
        return "ingreso medio-alto"
    else:
        return "ingreso alto"

# Una vez definidas, se reutilizan cuantas veces haga falta
ecuador = pib_per_capita(121_000_000_000, 18_000_000)
print(f"Ecuador: {ecuador:,.0f} USD → {clasificar_ingreso(ecuador)}")

chile = pib_per_capita(335_000_000_000, 20_000_000)
print(f"Chile:   {chile:,.0f} USD → {clasificar_ingreso(chile)}")
````

### 🔨 Tú haces 6

1. Define `promedio(lista)` que **devuelva** el promedio de una lista de números (reutiliza tu acumulador del bloque 5).
2. Define `es_par(n)` que devuelva `True` si `n` es par y `False` si es impar (una sola línea con `return`).
3. Prueba ambas: `promedio(notas)` debe dar `8.2`; `es_par(10)` → `True`, `es_par(7)` → `False`.

````python
# Tu código aquí
# def promedio(lista):
#     ...
# def es_par(n):
#     ...
````

#### ✅ Solución

````python
def promedio(lista):
    """Devuelve el promedio de una lista de números."""
    suma = 0
    for valor in lista:
        suma += valor
    return suma / len(lista)

def es_par(n):
    """Devuelve True si n es par, False si es impar."""
    return n % 2 == 0

print(promedio(notas))        # 8.2
print(es_par(10), es_par(7))  # True False
````
